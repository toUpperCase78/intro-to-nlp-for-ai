# Intro to NLP for AI | 03 - Identifying Parts of Speech and Named Entities | Practical Example

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

### Load Data

```python
bbc = pd.read_csv('bbc_news.csv')
```

```python
bbc.info()
```

```python
bbc.head()
```

```python
bbc.loc[0]
```

```python
titles = pd.DataFrame(bbc['title'])
```

```python
titles
```

### Clean Title Data

```python
# LOWERCASE
titles['lowercase'] = titles['title'].str.lower()
titles.head()
```

```python
# REMOVE STOPWORDS
en_stopwords = stopwords.words('english')
titles['no_stopwords'] = titles['lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
# REMOVE PUNCTUAITONS
titles['no_stopwords_no_punct'] = titles.apply(lambda x: re.sub(r"[^\w\s]", "", x['no_stopwords']), axis=1)
```

```python
# TOKENIZATION
titles['tokens_raw'] = titles.apply(lambda x: word_tokenize(x['title']), axis=1)
titles['tokens_clean'] = titles.apply(lambda x: word_tokenize(x['no_stopwords_no_punct']), axis=1)
```

```python
# LEMMATIZATION
lemmatizer = WordNetLemmatizer()
titles['tokens_clean_lemmatized'] = titles['tokens_clean'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
titles.head()
```

```python
tokens_raw_list = sum(titles['tokens_raw'], [])
tokens_clean_list = sum(titles['tokens_clean_lemmatized'], [])
```

```python
# print(tokens_clean_list)
```

### POS Tagging

```python
nlp = spacy.load('en_core_web_sm')
```

```python
# POS TAGGING ON RAW TEXT
spacy_doc = nlp(' '.join(tokens_raw_list))
```

```python
pos_df = pd.DataFrame(columns=['tokens', 'pos_tag'])
```

```python
for token in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{'tokens': token.text, 'pos_tag': token.pos_}])], ignore_index=True)
```

```python
pos_df.head(20)
```

```python
pos_df_counts = pos_df.groupby(['tokens', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
pos_df_counts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts['pos_tag'] == 'NOUN'][:10]
nouns
```

```python
verbs = pos_df_counts[pos_df_counts['pos_tag'] == 'NOUN'][:10]
verbs
```

```python
adjectives = pos_df_counts[pos_df_counts['pos_tag'] == 'ADJ'][:10]
adjectives
```

```python
# POS TAGGING ON CLEAN TEXT
spacy_doc_clean = nlp(' '.join(tokens_clean_list))
```

```python
pos_df_clean = pd.DataFrame(columns=['tokens', 'pos_tag'])
for token in spacy_doc_clean:
    pos_df_clean = pd.concat([pos_df_clean, pd.DataFrame.from_records([{'tokens': token.text, 'pos_tag': token.pos_}])], ignore_index=True)
```

```python
pos_df_clean.head(15)
```

```python
pos_df_clean_counts = pos_df_clean.groupby(['tokens', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
pos_df_clean_counts.head(10)
```

```python
nouns_clean = pos_df_clean_counts[pos_df_clean_counts['pos_tag'] == 'NOUN'][:10]
nouns_clean
```

### Named Entity Recognition (NER)

```python
ner_df = pd.DataFrame(columns=['token', 'ner_tag'])
for token in spacy_doc.ents:
    if pd.isna(token.label_) is False:
        ner_df = pd.concat([ner_df, pd.DataFrame.from_records([{'token': token.text, 'ner_tag': token.label_}])], ignore_index=True)
```

```python
ner_df.head(20)
```

```python
ner_df_counts = ner_df.groupby(['token', 'ner_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
ner_df_counts.head(10)
```
