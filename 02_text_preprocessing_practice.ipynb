# Intro to NLP for AI | 02 - Text Preprocessing | Practical Example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('tripadvisor_hotel_reviews.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.loc[0, 'Review']
```

```python
df['review_lowercase'] = df['Review'].str.lower()
```

```python
df.head()
```

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove('not')
```

```python
print(en_stopwords)
```

```python
df['review_no_stopwords'] = df['Review'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
df.head()
```

```python
df.loc[0,'Review']
```

```python
df.loc[0,'review_lowercase']
```

```python
df.loc[0,'review_no_stopwords']
```

```python
df['review_no_stopwords_no_punct'] = df.apply(lambda x: re.sub(r"[*]", "star", x['review_no_stopwords']), axis=1)
```

```python
df['review_no_stopwords_no_punct'] = df.apply(lambda x: re.sub(r"[^\w\s]", "", x['review_no_stopwords_no_punct']), axis=1)
```

```python
df.head()
```

```python
df['review_tokenized'] = df.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punct']), axis=1)
```

```python
print(df.loc[0, 'review_tokenized'])
```

```python
ps = PorterStemmer()
df['review_stemmed'] = df['review_tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
df['review_lemmatized'] = df['review_tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
df[['review_tokenized', 'review_stemmed', 'review_lemmatized']].head()
```

```python
print(df.loc[0, 'review_stemmed'])
```

```python
print(df.loc[0, 'review_lemmatized'])
```

```python
tokens_together = sum(df['review_lemmatized'], [])
```

```python
len(tokens_together)
```

```python
unigrams = pd.Series(nltk.ngrams(tokens_together, 1)).value_counts()
print(unigrams)
```

```python
bigrams = pd.Series(nltk.ngrams(tokens_together, 2)).value_counts()
print(bigrams)
```

```python
trigrams = pd.Series(nltk.ngrams(tokens_together, 3)).value_counts()
print(trigrams)
```
